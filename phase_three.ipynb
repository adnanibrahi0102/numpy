### Advance operation with business example

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Dataset for the sales data id , sales of 2021 , sales of 2020 , sales of 2019 , sales of 2018
from numpy import mean


sales_data = np.array([
    [1 , 150000 , 120000 , 170000 , 200000],
    [2 , 200000 , 180000 , 220000 , 250000],
    [3 , 250000 , 220000 , 270000 , 300000],
    [4 , 300000 , 260000 , 310000 , 340000],
    [5 , 350000 , 300000 , 350000 , 380000],
])

print("Shape of the sales data : " , sales_data.shape)
print("sales data of first three restaurants :")
print(sales_data[0:3])

# Extracting the sales data of 2021 exclude the id column
sales_2021 = sales_data[:,1]
print("sales data of 2021 :" , sales_2021)

# max sales of restaurants
max_sales = np.max(sales_data[:,1:], axis=0)
print("max sales of restaurants : " , max_sales)

# min sales of restaurants
min_sales = np.min(sales_data[:,1:], axis=0)
print("min sales of restaurants : " , min_sales)

# min sales of resturants in 2022
min_sales_2018 = np.min(sales_data[:,-1], axis=0)
print("min sales of restaurants in 2018 : " , min_sales_2018)

# average sales of restaurants
average_sales =np.mean(sales_data[: , 1:], axis=0)
print("average sales of restaurants : " , average_sales)

# cummulative sales of restaurants means sum of all sales of restaurants
cummulative_sales = np.sum(sales_data[: , 1:], axis=0)
print("cummulative sales of restaurants : " , cummulative_sales)
# Sales trend graph
years = [2018, 2019, 2020, 2021]
plt.figure(figsize=(8, 4))
plt.plot(years, average_sales, marker='o', linestyle='-', color='b', label="Avg Sales")
plt.title("Average Sales of Restaurants Over Years")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.xticks(years)  # Ensure correct year labels
plt.grid(True)
plt.legend()
plt.show()

```
